## connecting the excel file


```python
import pandas as pd
import openpyxl
df =pd.read_excel("raw_data.xlsx")
print(df.head())
print(df.tail())
print(df.info())
print(df.isnull().sum())
missing = df.isnull().sum()
print("missing columns : ",missing[missing>0])
rawdf  = df.copy()
```

### handling missing values

```python
missing_percent = (df.isnull().sum()/len(df))*100
missing_percent = missing_percent[missing_percent>0]
print(" column with missing values and its percentage: ",missing_percent)
print("\n")
df['CouponCode'] = df['CouponCode'].fillna("No coupon")
print(df.isnull().sum())
```

### duplicate checking

```python
print("duplicate rows:", df.duplicated().sum())
print("duplicate order IDs : ",df["OrderID"].duplicated().sum())
```

### standardize text columns 

```python
text_col = [ "Product","ShippingAddress",
             "PaymentMethod", "OrderStatus","CouponCode",
             "ReferralSource" ]
for col in text_col:
    df[col]=df[col].astype(str).str.strip().str.title()
print(df.head())
```

### Date formatting

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Date"]= df["Date"].dt.strftime("%Y-%m-%d")
print(df.head())
```

### Numeric Precision

```python
df["UnitPrice"]=df["UnitPrice"].round(2)
df["TotalPrice"]=df["TotalPrice"].round(2)
print(df.head())
```

### Double checking missing values 

```python
fake_missing = [
    "N/A",
    "NA",
    " ",
    "Null",
    "NULL",
    "-"," ","",
    "none",
    "None","nan","NaN"]
for col in df.columns:
    count=df[col].astype(str).isin(fake_missing).sum()
    if count>0:
        print(f"{col}:{count}")
```

### checking weird category

```python
print(df['PaymentMethod'].unique())
print(df['OrderStatus'].unique())
print(df['Product'].unique())
print(df['ReferralSource'].unique())
```

### checking impossible values

```python
print(df[df['Quantity']<=0])
```

### validating total price

```python
invalid_total = df[(df['Quantity']*df['UnitPrice']).round(2)!=df["TotalPrice"].round(2)]
print(invalid_total)
```

### checking outliers

```python
print(df['TotalPrice'].describe())
```

```python
Q1=df['TotalPrice'].quantile(0.25)
Q3=df['TotalPrice'].quantile(0.75)
IQR =Q3-Q1
LOWER = Q1-1.5 *IQR
UPPER = Q3+1.5*IQR
OUTLIER = df[(df['TotalPrice']<LOWER) |(df['TotalPrice']>UPPER)]
print(OUTLIER[['OrderID',"TotalPrice"]])
```

### Create raw copy

```python
raw_df  = df.copy()
```

### SUMMARY

```python
summary = pd.DataFrame({ "Metric": ['rows before cleaning',
                                    'rows after cleaning',
                                    'missing values handled',
                                    'duplicate rows','duplicate order IDs',
                                    'invalid quantity rows','pricing errors','Outliers detected'],
                         "value": [ len(raw_df),len(df),309,0,0,0,0,len(OUTLIER)]
                       })
print(summary)
```
